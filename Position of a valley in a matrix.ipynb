```python
row=int(input("\nEnter the number of rows: "))
col=int(input("Enter the number of columns: "))

print("Enter the elements of the matrix: ")
mat=[[int(input()) for j in range(col)] for i in range(row)]

for i in range(row):
    for j in range(col):
        print(mat[i][j], end=' ')
    print()

    
count=0
valley=[]
for i in range(row):
    for j in range(col):
        left=True
        right=True
        up=True
        down=True
        for k in range(0,j):
            if mat[i][k] <= mat[i][k+1]:
                left=False
                break
        for k in range(j,col-1):
            if mat[i][k] >= mat[i][k+1]:
                right=False
                break
        for k in range(0,i):
            if mat[k][j] <= mat[k+1][j]:
                up=False
                break
        for k in range(i,row-1):
            if mat[k][j] >= mat[k+1][j]:
                down=False
                break
        if (left+right+up+down) == 4:
            valley.append((i,j))
            count += 1

print()            
if len(valley) == 0:
    print("There is no valley.")
elif len(valley) == 1:
    print("The position of the valley is",valley[0])
else:
    print("There are {} valleys and the positions are".format(len(valley)))
    for position in valley:
        print(position)
            
```
